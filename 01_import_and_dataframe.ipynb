## Validaciones de los valores del df

Validamos y rellenamos valores en nuestro dataframe

````python
pd.isnull(data["body"]) # Son null
pd.notnull(data["body"]) # No son null

data.dropna(axis=0, how="all")
# all: si todos son Nan, borro la fila
# any: si uno es Nan, borro la fila

data.fillna(0) # Relleno Nan por cero
#data.fillna("Desconocido") # Relleno Nan por Desconocido
````

````python
# relleno los Nan con el valor anterior
data['body'].fillna(method="ffill")

# relleno los Nan con el valor siguiente
data['body'].fillna(method="backfill")

# relleno con la media
data['body'].fillna(data['body'].mean())
````

````python
# genero una columna por cada valor de la columna sexo, con cero o uno
dummy_sex = pd.get_dummies(data["sex"], prefix="sex") 

# data = data.drop(["sex"], axis = 1) elimino columna
dummy_sex
````

## Concatenar

con **concat** podemos concatenar mas de 1 dataframe.

Con **axis** podemos especificar si queremos concatenar como filas o columas

````python
data = pd.concat([data, dummy_sex], axis = 1)
data
````
